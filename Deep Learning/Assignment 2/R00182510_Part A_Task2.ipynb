```python

import tensorflow as tf
print(tf.__version__)

import numpy as np
import h5py
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from keras import regularizers
from sklearn.svm import SVC
```

```python
def loadDataH5():
        

    with h5py.File('/home/jupyter/dataset/data1.h5','r') as hf:
        trainX = np.array(hf.get('trainX'))
        trainY = np.array(hf.get('trainY'))
        valX = np.array(hf.get('valX'))
        valY = np.array(hf.get('valY'))
        print (trainX.shape,trainY.shape)
        print (valX.shape,valY.shape)
    return trainX, trainY, valX, valY
```

```python
def plotAccLoss(H, NUM_EPOCHS):

    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["loss"], label="train_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["accuracy"], label="train_acc")
    plt.plot(np.arange(0, NUM_EPOCHS), H.history["val_accuracy"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()
    plt.show()
```

## Basic Neural network Ensemble

```python
def cnnModel(width=28, height=28, depth=1, classes=10):
  
    # initialize the model
    model = keras.models.Sequential()

    inputShape = (height, width, depth)


    # first CONV => CONV => POOL layer set
    model.add(keras.layers.Conv2D(32, (3, 3), padding="same", input_shape=inputShape, activation='relu'))
    model.add(keras.layers.Conv2D(32, (3, 3), padding="same",activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(keras.layers.SpatialDropout2D(rate=0.5))

    # second CONV => CONV => POOL layer set
    model.add(keras.layers.Conv2D(64, (3, 3), padding="same",activation='relu'))
    model.add(keras.layers.Conv2D(64, (3, 3), padding="same",activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(keras.layers.SpatialDropout2D(rate=0.5))

    # first (and only) set of FC => RELU layers
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(512,activation='relu'))
    model.add(keras.layers.Dropout(rate=0.5))

    # softmax classifier
    model.add(keras.layers.Dense(classes, activation='softmax'))

    # return the constructed network architecture
    return model


def main():
  # resize all images to thie width and height
    NUM_EPOCHS = 50
    trainX, trainY, testX, testY = loadDataH5()
    width = trainX.shape[1]
    height = trainX.shape[2]
    depth = trainX.shape[3]
    batchSize = 16
    all_learner_probabilites = []
    
    for i in range(5):
    
        print("Compiling model "+str(i+1)+"...")
        opt = keras.optimizers.SGD(lr=0.01)
        model = cnnModel(width=width, height=height, depth=depth, classes=17)
        model.compile(loss="sparse_categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
#         print (model.summary())

        trainDataGenerator = tf.keras.preprocessing.image.ImageDataGenerator(
        shear_range=0.2,
        zoom_range=0.2,
        rotation_range=30,
        horizontal_flip=True)

        train_generator = trainDataGenerator.flow(trainX, trainY, batch_size=16)

        fname= "checkpointed_model_"+str(i+1)+".hdf5"
        checkpoint = tf.keras.callbacks.ModelCheckpoint(fname, monitor="val_loss", mode="min", save_best_only=True, verbose=1)
        
        print("Training network"+str(i)+"...")
        history = model.fit(train_generator, validation_data = (testX, testY),
                              steps_per_epoch=len(trainX)/ 16, epochs=NUM_EPOCHS, callbacks=[checkpoint])
        
        
        model.load_weights("checkpointed_model_"+str(i+1)+".hdf5")
#         model.compile(loss="sparse_categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
        
        results = model.predict(testX)
        val_preds = np.argmax(results, axis=1)
        print("\nModel accuracy on Test Set",accuracy_score(testY, val_preds))
        
        plotAccLoss(history, NUM_EPOCHS)
      
        all_learner_probabilites.append(results)
        
    all_learner_probabilites = np.array(all_learner_probabilites)

    accum_probabilities = all_learner_probabilites.sum(axis = 0)/17
    accum_preds = np.argmax(accum_probabilities, axis=1)
    
    print("\nFinal ensemble network accuracy on Test Set",accuracy_score(testY, accum_preds))
    
        
main()
    
```

## A more complex neural network Ensemble

```python
# Feature extraction using Xception
def Xception_feature_extraction(trainX, trainY, testX, testY):
    width = trainX.shape[1]
    height = trainX.shape[2]
    depth = trainX.shape[3]
    
    # set up the Xception model for feature extarction
    XceptionModel = tf.keras.applications.Xception(weights='imagenet', include_top=False, input_shape=(width, height, depth))

    featuresTrain = XceptionModel.predict(trainX)
    featuresTrain = featuresTrain.reshape(featuresTrain.shape[0], -1)
    featuresTest = XceptionModel.predict(testX)
    featuresVal = featuresTest.reshape(featuresTest.shape[0], -1)
    
    # from sklearn.svm import SVC
    model = SVC(decision_function_shape='ovr', probability = True)
    # train the model
    H = model.fit(featuresTrain, trainY)
    # evaluate test data
    results = model.predict_proba(featuresVal)
    val_preds = np.argmax(results, axis=1)

    print ("Validation Accuracy from Model 1: ", metrics.accuracy_score(val_preds, testY))
    # return the probabilities
    return results
```

```python
# Feature extraction using VGG16
def VGG16_feature_extraction(trainX, trainY, testX, testY):
    width = trainX.shape[1]
    height = trainX.shape[2]
    depth = trainX.shape[3]
    # set up the VGG16 model for feature extarction
    VGG16Model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(width, height, depth))

    featuresTrain = VGG16Model.predict(trainX)
    featuresTrain = featuresTrain.reshape(featuresTrain.shape[0], -1)
    featuresTest = VGG16Model.predict(testX)
    featuresVal = featuresTest.reshape(featuresTest.shape[0], -1)
    # from sklearn.svm import SVC
    model = SVC(decision_function_shape='ovr', probability = True)
    # train the model
    H = model.fit(featuresTrain, trainY)
    # evaluate test data
    results = model.predict_proba(featuresVal)
    val_preds = np.argmax(results, axis=1)

    print ("Validation Accuracy from Model 2: ", metrics.accuracy_score(val_preds, testY))
    # return the probabilities
    return results
```

```python
# Fien-tune using Xception
def Xception_fine_tuned(trainX, trainY,testX, testY, NUM_EPOCHS, batchSize):
    
    width = trainX.shape[1]
    height = trainX.shape[2]
    depth = trainX.shape[3]
    opt = keras.optimizers.SGD(lr=0.01)
    
    # set up the Xception model for fine-tuning with trainable as False
    XceptionModel = tf.keras.applications.Xception(weights='imagenet', include_top=False, input_shape=(height, width, depth))

    XceptionModel.trainable = False
    model = tf.keras.models.Sequential()
    model.add(XceptionModel)
    model.add(tf.keras.layers.Flatten())
    # include fully connected layers
    model.add(tf.keras.layers.Dense(256, activation='relu', kernel_regularizer = regularizers.l2(0.05)))  
    model.add(tf.keras.layers.Dropout(0.5))
    model.add(tf.keras.layers.Dense(17, activation='softmax'))
    print (model.summary())
    
    # Introduce data augmentation with ImageDataGenerator
    trainDataGenerator = tf.keras.preprocessing.image.ImageDataGenerator(
      shear_range=0.2,
      zoom_range=0.2,
      rotation_range=30,
      horizontal_flip=True)
    
    train_generator = trainDataGenerator.flow(trainX, trainY, batch_size=batchSize)

    print ("\n Phase A - Training Fully Connected Layers\n")
    model.compile(loss='sparse_categorical_crossentropy', optimizer=tf.keras.optimizers.RMSprop(lr=0.001),
              metrics=['accuracy'])
    # train the Phase A model
    H =model.fit(train_generator, 
           epochs=NUM_EPOCHS, steps_per_epoch=len(trainX)/ batchSize, validation_data = (testX,testY))

    plotAccLoss(H, NUM_EPOCHS)
    
    # set the trainable flag as True for layers below block12_sepconv1
    print ("\n Phase B  - Fine Tune Fully Connected Layer and Selected Convolutional Layers \n")
    XceptionModel.trainable = True
    trainableFlag = False
    for layer in XceptionModel.layers:
        if layer.name == 'block12_sepconv1':
            trainableFlag = True
        layer.trainable = trainableFlag
    model.compile(loss='sparse_categorical_crossentropy',
                optimizer=tf.keras.optimizers.RMSprop(lr=1e-5),
                metrics=['accuracy'])
    
    # introduce the callbacks for checkpointing
    fname= "checkpointed_model.hdf5"
    checkpoint = tf.keras.callbacks.ModelCheckpoint(fname, monitor="val_loss", mode="min", save_best_only=True, verbose=1)
    
    # re-train the model - phase B
    H =model.fit(train_generator, epochs=NUM_EPOCHS, steps_per_epoch=len(trainX)/ batchSize, validation_data = (testX,testY),
                 callbacks=[checkpoint])
    
    # load the checkpointed model weights
    model.load_weights("checkpointed_model.hdf5")
    
    # evaluate the model
    results = model.predict(testX)
    val_preds = np.argmax(results, axis=1)
    print("\nValidation Accuracy from Model 3: ",accuracy_score(testY, val_preds))    

    plotAccLoss(H, NUM_EPOCHS)
    # return the probabilities
    return results
```

```python
def VGG16_fine_tuned(trainX, trainY,testX, testY, NUM_EPOCHS, batchSize):
    
    width = trainX.shape[1]
    height = trainX.shape[2]
    depth = trainX.shape[3]
    opt = keras.optimizers.SGD(lr=0.01)
    
    # setup the VGG16 model with trainable as False for all top layers
    vggModel = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(height, width, depth))

    vggModel.trainable = False
    model = tf.keras.models.Sequential()
    model.add(vggModel)
    # include fully connected layers
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(256, activation='relu', kernel_regularizer = regularizers.l2(0.05)))  
    model.add(tf.keras.layers.Dropout(0.5))
    model.add(tf.keras.layers.Dense(17, activation='softmax'))
    print (model.summary())
    
    # Introduce data augmentation with ImageDataGenerator
    trainDataGenerator = tf.keras.preprocessing.image.ImageDataGenerator(
      shear_range=0.2,
      zoom_range=0.2,
      rotation_range=30,
      horizontal_flip=True)
    
    train_generator = trainDataGenerator.flow(trainX, trainY, batch_size=batchSize)


    print ("\n Phase A - Training Fully Connected Layers\n")
    model.compile(loss='sparse_categorical_crossentropy', optimizer=tf.keras.optimizers.RMSprop(lr=0.001),
              metrics=['accuracy'])
    # train the Phase A model
    H =model.fit(train_generator, 
           epochs=NUM_EPOCHS, steps_per_epoch=len(trainX)/ batchSize, validation_data = (testX,testY))

    plotAccLoss(H, NUM_EPOCHS)

    # set the trainable flag as True for layers below block4_conv1
    print ("\n Phase B  - Fine Tune Fully Connected Layer and Selected Convolutional Layers \n")
    vggModel.trainable = True
    trainableFlag = False
    for layer in vggModel.layers:
        if layer.name == 'block4_conv1':
            trainableFlag = True
        layer.trainable = trainableFlag
    model.compile(loss='sparse_categorical_crossentropy',
                optimizer=tf.keras.optimizers.RMSprop(lr=1e-5),
                metrics=['accuracy'])
    
    # introduce the callbacks for checkpointing and learning rate scheduler
    fname= "checkpointed_model.hdf5"
    checkpoint = tf.keras.callbacks.ModelCheckpoint(fname, monitor="val_loss", mode="min", save_best_only=True, verbose=1)
    # re-train the model - phase B
    H =model.fit(train_generator, epochs=NUM_EPOCHS, steps_per_epoch=len(trainX)/ batchSize, validation_data = (testX,testY),
                 callbacks=[checkpoint])
    
    # load the checkpointed model weights
    model.load_weights("checkpointed_model.hdf5")
    # evaluate the model
    results = model.predict(testX)
    val_preds = np.argmax(results, axis=1)
    print("\nValidation Accuracy from Model 4: ",accuracy_score(testY, val_preds))    

    plotAccLoss(H, NUM_EPOCHS)
    # return the probabilities
    return results
```

```python
# Feature extraction using Xception and VGG16
def Xception_VGG16_feature_extraction(trainX, trainY, testX, testY):

    width = trainX.shape[1]
    height = trainX.shape[2]
    depth = trainX.shape[3]
    
    # set up the Xception model for feature extraction
    XceptionModel = tf.keras.applications.Xception(weights='imagenet', include_top=False, input_shape=(width, height, depth))
    # set up the VGG16 model for feature extarction
    VGGModel = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(width, height, depth))
    
    # extract features fro train and test from both the models
    featuresTrain1 = XceptionModel.predict(trainX)
    featuresTrain1 = featuresTrain1.reshape(featuresTrain1.shape[0], -1)
    featuresTest1 = XceptionModel.predict(testX)
    featuresVal1 = featuresTest1.reshape(featuresTest1.shape[0], -1)

    featuresTrain2 = VGGModel.predict(trainX)
    featuresTrain2 = featuresTrain2.reshape(featuresTrain2.shape[0], -1)
    featuresTest2 = VGGModel.predict(testX)
    featuresVal2 = featuresTest2.reshape(featuresTest2.shape[0], -1)

    print(featuresTrain1.shape)
    print(featuresTrain2.shape)
    
    # stack the features from both the models
    features_all = np.hstack([featuresTrain1,featuresTrain2])
    labels_all = np.hstack([trainY, trainY])

    # from sklearn.svm import SVC
    from sklearn.svm import SVC
    model = SVC(decision_function_shape='ovr', probability=True)
    # train the model
    H = model.fit(features_all, trainY)
    # evaluate test data
    results = model.predict_proba(np.hstack([featuresVal1, featuresVal2]))
    print(results.shape)
    val_preds = np.argmax(results, axis=1)

    print ("Validation Accuracy from Model 1: ", metrics.accuracy_score(val_preds, testY))
    # return the probabilities
    return results
```

```python
def main():
    NUM_EPOCHS = 50
    # Load input data
    trainX, trainY, testX, testY = loadDataH5()
    batchSize = 16
    all_learner_probabilites = []
    
    # Execute the five different base learners and record their prediction probabilities
    for i in range(5):
    
        print("Executing model "+str(i+1)+"...")
        
        if i == 0:
            results = Xception_feature_extraction(trainX, trainY, testX, testY)
            all_learner_probabilites.append(results)
        elif i == 1:
            results = VGG16_feature_extraction(trainX, trainY, testX, testY)
            all_learner_probabilites.append(results)
        elif i == 2:
            results = Xception_fine_tuned(trainX, trainY,testX, testY, NUM_EPOCHS, batchSize)
            all_learner_probabilites.append(results)
        elif i == 3:
            results = VGG16_fine_tuned(trainX, trainY,testX, testY, NUM_EPOCHS, batchSize)
            all_learner_probabilites.append(results)
        elif i == 4:
            results = Xception_VGG16_feature_extraction(trainX, trainY, testX, testY)
            all_learner_probabilites.append(results)
        else:
            print('Unknown ensemble')
        
    # Combine the probabilites from the base learners as an np array
    all_learner_probabilites = np.array(all_learner_probabilites)
    
    # average the probabilities from all the base learners and determine the target class
    accum_probabilities = all_learner_probabilites.sum(axis = 0)/17
    accum_preds = np.argmax(accum_probabilities, axis=1)
    
    # calculate final ensemble accuracy
    print("\nFinal ensemble network accuracy on Test Set",accuracy_score(testY, accum_preds))
    
        
main()
    
```
